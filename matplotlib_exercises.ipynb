```python
import matplotlib.pyplot as plt
```

```python
#Use matplotlib to plot the following equation: y = x^2 - x + 2
x=[]
y=[]
x =list(range(-20,20))
y = [(x**2 - x + 2) for x in x]
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,y)
plt.ylim(-5,30)
plt.xlim(-10,10)
plt.hlines(0,-10,30, ls="--")
plt.vlines(0,-5,50, ls="--")

plt.title("graph of function")
plt.xlabel("x axis")
plt.ylabel("y axis")
#,loc="upper right")

plt.annotate(
            "this is origin", 
            xy=(0,0), 
            xytext=(3,-4), 
            arrowprops=({
                        "facecolor":"black",
                        })
            )

```

Create and label 4 separate charts for the following equations (choose a range for x that makes sense): \
y=âˆšx \
y=x^3 \
y=2^x \
y=1/(x+1)

```python
plt.figure(figsize=(12,12))
plt.suptitle("graphs")

x1 =list(range(0,64))
y1 = [(ele**.5) for ele in x1] #y=âˆšx
plt.subplot(221)
plt.plot(x1,y1)
plt.ylim(-1,8)
plt.xlim(-1,8)
plt.hlines(0,-1,20, ls="--")
plt.vlines(0,-1,20, ls="--")
plt.title("y=x^.5")

x2 =list(range(-4,4))
y2 = [(ele**3) for ele in x2] #y=x^3
plt.subplot(222)
plt.plot(x2,y2)
plt.ylim(-20,20)
plt.xlim(-10,10)
plt.hlines(0,-10,20, ls="--")
plt.vlines(0,-20,40, ls="--")
plt.title("y=x^3")

x3 =list(range(-10,10))
y3 = [(2**ele) for ele in x3]#y=2^x
plt.subplot(223)
plt.plot(x3,y3)
plt.ylim(-5,30)
plt.xlim(-10,10)
plt.hlines(0,-10,30, ls="--")
plt.vlines(0,-5,50, ls="--")
plt.title("y=x^2")

x4 =list(range(-10,20))
y4 = [(1/(ele + 1)) if (ele != -1) else (1/(ele + 1.001)) for ele in x4] #y=1/(x+1)
plt.subplot(224)
plt.plot(x4,y4)
plt.ylim(-.5,2)
plt.xlim(-1,10)
plt.hlines(0,-.5,30, ls="--")
plt.vlines(0,-1,50, ls="--")
plt.title("y=1/(x+1)")



plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.suptitle("graphs")

x1 =list(range(0,64))
y1 = [(ele**.5) for ele in x1] #y=âˆšx
plt.plot(x1,y1,label = "y=x^.5")

x2 =list(range(-4,4))
y2 = [(ele**3) for ele in x2] #y=x^3
plt.plot(x2,y2,label = "y=x^3")

x3 =list(range(-10,10))
y3 = [(2**ele) for ele in x3]#y=2^x
plt.plot(x3,y3,label = "y=x^2")

x4 =list(range(-10,20))
y4 = [(1/(ele + 1)) if (ele != -1) else (1/(ele + 1.001)) for ele in x4] #y=1/(x+1)
plt.plot(x4,y4,label = "y=1/(x+1)")

plt.ylim(-10,10)
plt.xlim(-10,10)
plt.hlines(0,-10,20, ls="--")
plt.vlines(0,-10,20, ls="--")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.show()
```
