```python
#Copy the code from the lesson to create a dataframe full of student grades.
import pandas as pd
import numpy as np

np.random.seed(1349)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
len(math_grades) == len(english_grades) == len(reading_grades) == len(students)
##verifies all lengths match
```

```python
df.info() #size
df.head() #display first few
df.describe() #basic descriptive statistics
```

```python
#Create a column named passing_english that 
# indicates whether each student has a passing grade in english.
df['passing_english'] = df.english > 70 #return added dataframe col of bool where english is above 70
df
```

```python
#Sort the english grades by the passing_english column. 
# How are duplicates handled?
df.sort_values(by=["passing_english","english"],ascending=False) #sort values first by passsing_english and then by english(grade) in descending order (true = 1 so it is descending down to 0/false)
#duplicates are typically sorted in ascending index order
```

```python
#Sort the english grades first by passing_english and then by student name. 
# All the students that are failing english should be first, 
# and within the students that are failing english they should 
# be ordered alphabetically. 
# The same should be true for the students passing english. 
# (Hint: you can pass a list to the .sort_values method)
df.sort_values(by=["passing_english","name"],ascending=True)
```

```python
#Sort the english grades first by passing_english, 
# and then by the actual english grade, 
# similar to how we did in the last step.
df.sort_values(by=["passing_english","english"],ascending=True)
```

```python
#Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.
df['overall_grade'] = ((df.english + df.math + df.reading)/3)
df
```

```python
#Rename the cty column to city.
#Rename the hwy column to highway.
df2.rename(columns={"cty":"city","hwy":"highway"},inplace=True)
df2
```

```python
#Do any cars have better city mileage than highway mileage?
df2_comparison = df2.apply(lambda x:x["city"] if x["city"] > x["highway"] else False,axis=1)
if df2_comparison.value_counts()[0] == len(df2_comparison.index):
    print("no")
else:
    print("maybe")
# mask = mpg.city > mpg.highway
# mpg[mask] ##empty
# len(mpg[mask])
# mask.sum()
# all equal 0
```

```python
#Create a column named mileage_difference this column should 
# contain the difference between highway and city mileage for each car.
df2["mileage_difference"] = (df2.highway - df2.city)
df2
```

```python
#Which car (or cars) has the highest mileage difference?
df2[df2.mileage_difference == df2.mileage_difference.max()]
```

```python
df2["class"].dtype, type(df2["class"])
```

```python
#Which compact class car has the lowest highway mileage?

compact_min = df2[df2["class"] == "compact"].mileage_difference.min()
df2[
    (df2["class"] == "compact") 
    & 
    (df2["mileage_difference"] == compact_min)
    ]
#df2[(df2["class"] == 'compact') & (df2(df2["class"] == 'compact').highway.max())]
```

```python
# The best?
compact_max = df2[df2["class"] == "compact"].mileage_difference.max()
df2[
    (df2["class"] == "compact") 
    & 
    (df2["mileage_difference"] == compact_max)
    ]
```

```python
#Create a column named average_mileage that is 
# the mean of the city and highway mileage.
df2["average_mileage"] = ((df2.highway + df2.city) / 2)
df2
# 2/ ((1/mpg.city) + (1/mpg.highway))
```

```python
#Which dodge car has the best average mileage? 
dodge_max = df2[df2["manufacturer"] == "dodge"].average_mileage.max()
df2[(df2["manufacturer"] == "dodge") & (df2["average_mileage"] == dodge_max)]
```

```python
# The worst?
dodge_min = df2[df2["manufacturer"] == "dodge"].average_mileage.min()
df2[(df2["manufacturer"] == "dodge") & (df2["average_mileage"] == dodge_min)]
```
